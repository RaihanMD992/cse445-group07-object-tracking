```python
from ultralytics import YOLO

# 1. Load the foundational "Pre-trained" model
# We use the 'nano' version (yolov8n) because it is incredibly fast and perfect for real-time video tracking.
model = YOLO('yolov8n.pt')

# 2. Train the model on your custom dataset
# We are setting epochs to 25 for a fast, MVP training run.
results = model.train(
    data='/content/dataset/data.yaml',
    epochs=25,
    imgsz=640,
    plots=True
)

print("Training Complete! Your custom weights are saved.")
```

```python
from ultralytics import YOLO
from IPython.display import display, Image

# 1. Load your brand new, custom-trained model
model = YOLO('/content/runs/detect/train/weights/best.pt')

# 2. Pick a test image (Copy the path of ONE image from your test folder)
test_image_path = '/content/dataset/test/images/images-39-_jpg.rf.0cadb4d606981a7db95ff2c01f4dbd78.jpg'

# 3. Run the model on the image
# save=True tells YOLO to draw the boxes and save the result
results = model.predict(source=test_image_path, save=True)

# 4. Display the result right here in Colab
# YOLO saves predictions in the 'runs/detect/predict/' folder
display(Image(filename=f'/content/runs/detect/predict/{test_image_path.split("/")[-1]}'))
```
