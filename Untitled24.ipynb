```python
from ultralytics import YOLO

# 1. Load the foundational "Pre-trained" model
# We use the 'nano' version (yolov8n) because it is incredibly fast and perfect for real-time video tracking.
model = YOLO('yolov8n.pt')

# 2. Train the model on your custom dataset
# We are setting epochs to 25 for a fast, MVP training run.
results = model.train(
    data='/content/dataset/data.yaml',
    epochs=25,
    imgsz=640,
    plots=True
)

print("Training Complete! Your custom weights are saved.")
```

```python
from ultralytics import YOLO
from IPython.display import display, Image
import glob
import os

# 1. Load your custom-trained model
model_path = '/content/runs/detect/train/weights/best.pt'
model = YOLO(model_path)

# 2. Define the path of your test image
test_image_path = '/content/dataset/test/images/images-45-2-_jpg.rf.db4afabf8ac32c0c5852848d38090b92.jpg'

# 3. Run the model on the image
# save=True will automatically create a new 'predict' folder
results = model.predict(source=test_image_path, save=True)

# 4. Automatically find and display the result
# Extract the filename to search for it in the dynamic folders
img_name = os.path.basename(test_image_path)

# Glob searches through all 'predict' folders to find your image
found_files = glob.glob(f'/content/runs/detect/*/{img_name}')

if found_files:
    print(f"Displaying result from: {found_files[0]}")
    display(Image(filename=found_files[0]))
else:
    print("Could not find the processed image. Check the 'runs/detect' folder in the sidebar.")
```
